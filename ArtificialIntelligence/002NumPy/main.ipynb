# 1.Numpy的属性介绍

```python

# numpy的数组 => ndarray， 他是多维的，并且是固定类型的，主要用于矩阵运算

# 导包

# import numpy as np

arr1 = np.arange(15).reshape(3,5)

print(arr1)
```

```python
print(f'arr1的维度: {arr1.ndim}')
print(f'arr1的轴数: {arr1.ndim}')
```

```python
print(f'arr1的形状: {arr1.shape}')
```

```python
print(f'arr1的元素类型: {arr1.dtype}')
```

```python
print(f'arr1的元素个数: {arr1.size}')
```

```python
# 一个字节占8位, 所以int64 / 8 = 8.
print(f'arr1的元素占用字节数: {arr1.itemsize}')
```

```python
print(f'arr1的类型: {type(arr1)}')
```

# 2.创建ndarray对象

### 2.1 通过np.array函数创建

```python
# 创建python列表
py_list = [11,22,33,44,55]
# 把python列表转为nbarray对象
arr2 = np.array(py_list)
```

```python
print(f'arr2 = {arr2}')
```

```python
print(f'arr2的类型 = {type(arr2)}')
```

### 2.2 通过np.arange函数创建

```python
# arange(起始, 结束, 步长, dtype=np.类型)
arr3 = np.arange(0,10,2,dtype=np.float64)
```

```python
print(f'arr3 = {arr3}')
```

```python
print(f'arr3的元素类型 = {arr3.dtype}')
```

### 2.3 随机数函数创建

```python
# np.random.rand(行，列) 生成 0.0 ~ 1.0 之间的随机小数矩阵
arr4 = np.random.rand(3,5)
```

```python
print(f'arr4 = {arr4}')
```

```python
# np.random.uniform(起始，结束，size=(行，列)) 生成指定范围的小数矩阵
arr5 = np.random.uniform(3,9,size=(2,6))
```

```python
print(f'arr5 = {arr5}')
```

# 3.类型转换

```python
# 创建ndarray对象
arr6 = np.arange(0,10,2,dtype=np.int32)
```

```python
# 把上述的array对象的元素类型从 int32 转换为 float32
arr7 = arr6.astype(np.float32)
```

```python
print(f'arr7 = {arr7}')
```

```python
print(f'arr6的元素类型 = {arr6.dtype}')
```

```python
print(f'arr7的元素类型 = {arr7.dtype}')
```

# 4.创建等比和等差数列

```python
# 等比数列
# np.logspace(起始指数，结束指数，元素长度，base=基数)
# 表示 2的0次幂, 2的9次幂，长度为10，基数为2的一个数组
arr8 = np.logspace(0,9,10,base=2)

print(f'arr8 = {arr8}')
```

```python
# 等差数列
# np.linspace(起始值，结束值，元素长度)
arr9 = np.linspace(1,10,10)
print(f'arr9 = {arr9}')
```
